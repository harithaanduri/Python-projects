6. Write a programming construct (create a user defined function) to calculate the median value of the data set wherever required.


```python
def med_val(df_husng):#defining function

  l = sorted(df_husng)#sorting values bfore calculating median
  n = len(l)#taking length of total data

  if n%2==0:# if lenth of the data is even
     x = l[n//2] #qufcnt value which is at middle of the data
     y = l[n//2 - 1]#before qufcnt value to middle value
     median = float((x + y)/2)#add both values and divide
     return median#return the value
  else: #if length of the total is odd
     median = (l[n+1//2])#divide by 2 to get middle value position
     return median #return median


print(med_val(df_husng['housing_median_age'])) #calling the function med_val by giving the data from dataset
```

10. Please create a new column named total_bedroom_size. If the total bedrooms is 10 or less, it should be quoted as small. If the total bedrooms is 11 or more but less than 1000, it should be medium, otherwise it should be considered large.



```python
df_husng['total_bedroom_size'] = 0 #creating a new column total_bedroom_size and value is 0.
df_husng.loc[df_husng['total_bedrooms']<=10,'total_bedroom_size'] = 'small' #giving if condition :If the total bedrooms is 10 or less, it should be quoted as small
df_husng.loc[df_husng['total_bedrooms']>=11,'total_bedroom_size'] = 'medium' #If condition:if the total bedrooms is >=11, it should be quoted as medium
df_husng.loc[df_husng['total_bedrooms']<1000,'total_bedroom_size'] = 'medium' #giving if condition :If the total bedrooms lessthan 1000  it should be quoted as medium
df_husng.loc[df_husng['total_bedrooms']>=1000,'total_bedroom_size'] = 'large' #giving if condition :If the total bedrooms is greater than 1000 it should be quoted as large
print(df_husng.head())
```

```python
df_husng['total_bedroom_size'] = 0
#creating empty column in df_husng
#using enumerate function to get the value at same index

for j,i in enumerate(df_husng['total_bedrooms']): #j is index,i is value

  if i <= 10:                                     #giving if condition :If the total bedrooms is 10 or less, it should be quoted as small,if true index,value go to the new column.
    df_husng.loc[j,'total_bedroom_size'] = 'small'
                                                 #using if condition based on totalbedrms , totalbedrmsize  encoded as small,medium,large
  elif 1000>i>=11:
    df_husng.loc[j,'total_bedroom_size'] = 'medium'
  else:
    df_husng.loc[j,'total_bedroom_size'] = 'large'

print(df_husng.head())
```
